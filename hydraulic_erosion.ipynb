## First Steps

Before I dive into the erosion model from the paper, I want to gather some libraries, figure out some data structures, and find a way to visualize the models as we go.

### Import Libraries

```python
import noise
import numpy as np
from matplotlib import pyplot as plt
```

### Make Some Noise

Before I dive into the erosion model, I want to start by generating a height map to erode. I'll put these noise values into a 2D numpy array.

```python
terrain_height = [
    [noise.snoise2(x/100, y/100) for x in range(256)]
    for y in range(256)
]

terrain_height = np.array(terrain_height)
```

### Displaying the Models

In this notebook I'll use matplotlib to show the arrays inline.

```python
plt.figure(figsize=(10,10))
plt.imshow(terrain_height)
plt.show()
```
